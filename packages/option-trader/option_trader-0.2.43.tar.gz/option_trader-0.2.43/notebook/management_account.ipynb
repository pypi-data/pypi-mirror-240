```python
user_name_list = mysite.get_user_list()

#print('{:10s}\t{:25s}\t{:10s}\t{:10s}\t{:10s}\t{:10s}'.format('user_name', 'account_name', 'cash', 'margin', 'asset', 'account') )    
#print('---------------------------------------------------------------------' )    

for user_name in user_name_list:
    u = mysite.get_user(user_name)
    account_name_list = u.get_account_list()
    for account_name in account_name_list:
        a = u.get_account(account_name)
        initial_balance = a.get_initial_balance()
        print('{:10s}\t{:25s}\t{:.2f}'.format(u.user_name, a.account_name, initial_balance))    
        print('{:10s}\t{:10s}\t{:10s}\t{:10s}\t{:10s}\t{:10s}'.format( 'account', 'asset','cash', 'margin', 'PL', 'gain') ) 
        acct_value = a.get_account_value()
        asset_value = a.get_asset_value()
        cash = a.get_cash_position()
        margin = a.get_margin_position()   
        pl = acct_value-initial_balance
        gain = 100 * pl / initial_balance
        print('{:.2f}\t{:10s}\t{:10s}\t{:10s}\t{:10s}\t{:.2f}%\r\n'.format(
            acct_value, str(round(asset_value,2)), str(round(cash,2)), str(round(margin,2)), str(round(pl,2)), gain))   
        
```

```python
from option_trader.admin import position_summary
from option_trader.admin import position
from option_trader.admin import transaction
from option_trader.admin import quote
from option_trader.consts import asset

acct = winwin_act 
opt = acct.get_position_summary()
pos = acct.get_positions()
trx = acct.get_transactions()
print(opt.shape[0], pos.shape[0], trx.shape[0], acct.get_initial_balance())
cash= round(acct.get_cash_position(),2)
margin = round(acct.get_margin_position(),2)
print(cash, margin)

open_pos = pos[pos[position.STATUS]==asset.OPENED] 

open_opt = opt[opt[position_summary.STATUS]==asset.OPENED] 
close_opt = opt[opt[position_summary.STATUS]==asset.CLOSED]

print(open_opt.shape[0], close_opt.shape[0])

close_opt.columns
```

```python
close_opt['days opened'] = close_opt.apply(lambda x: (pd.Timestamp(x[position_summary.STOP_DATE]).date()-pd.Timestamp(x[position_summary.TRADE_DATE]).date()).days, axis = 1)  

close_opt['days to expire'] = close_opt.apply(lambda x: (pd.Timestamp(x[position_summary.EXP_DATE]).date()-pd.Timestamp(x[position_summary.TRADE_DATE]).date()).days, axis = 1)  

close_opt['stop trigger'] = close_opt.apply(lambda x: (x['stop_reason'][0:9]), axis = 1) 

close_opt['hit target'] = close_opt.apply(lambda x: (x['last_stock_price'] >= x['target_low']) & (x['last_stock_price'] <= x['target_high']), axis=1)

close_opt['under target'] = close_opt.apply(lambda x: (x['last_stock_price'] < x['target_low']), axis=1)

close_opt['above target'] = close_opt.apply(lambda x: (x['last_stock_price'] > x['target_high']), axis=1)
```
