```python
import pandas as pd
import numpy as np
```

```python
# Reading the CSV file
original_data = pd.read_csv("./originales_clase_2/tb_sellout_01.txt.gz", header=None)
original_categories = pd.read_csv("./maestro_productos.csv")
```

```python
original_categories
```

```python
# print only rows of duplicated product_id
original_categories[original_categories.duplicated(['product_id'], keep=False)]
```

```python
#remove duplicates from original_categories
original_categories = original_categories.drop_duplicates(subset=['product_id'])
original_categories
```

```python
original_categories.to_csv("./maestro_productos.csv", index=False)
```

```python
data_series = original_data.iloc[:, 0]
column_names = data_series[:6].tolist()
data_series = data_series[6:]

# Ensuring the data length is a multiple of 5
data_series = data_series[: len(data_series) - (len(data_series) % 6)]

# Reshaping the data
reshaped_data = data_series.values.reshape(-1, 6)

df = pd.DataFrame(
    reshaped_data,
    columns=column_names,
)
```

```python
df
```

```python
df['periodo'] = df['periodo'].astype(int)
df['product_id'] = df['product_id'].astype(int)
df['customer_id'] = df['customer_id'].astype(int)
df['cust_request_qty'] = df['cust_request_qty'].astype(float)
df['cust_request_tn'] = df['cust_request_tn'].astype(float)
df['tn'] = df['tn'].astype(float)
```

```python
print(df["cust_request_qty"].sum(), df["cust_request_tn"].sum(), df["tn"].sum())
```

```python
# print unique number of customer_id, product_id, and period
print(df["customer_id"].nunique(), df["product_id"].nunique(), df["periodo"].nunique())
```

```python
df
```

```python
df_clientes = df.groupby(['customer_id', 'periodo'])[['cust_request_qty', 'cust_request_tn', 'tn']].sum()
df_productos = df.groupby(['product_id', 'periodo'])[['cust_request_qty', 'cust_request_tn', 'tn']].sum()
```

```python
print("ORIGINAL", df["cust_request_qty"].sum(), df["cust_request_tn"].sum(), df["tn"].sum())
print("CLIENTES", df_clientes["cust_request_qty"].sum(), df_clientes["cust_request_tn"].sum(), df_clientes["tn"].sum())
print("PRODUCTOS", df_productos["cust_request_qty"].sum(), df_productos["cust_request_tn"].sum(), df_productos["tn"].sum())
```

```python
df_productos.reset_index(inplace=True)
```

```python
df_productos
```

```python
df_productos = pd.merge(df_productos, original_categories[['product_id', 'cat1']], on='product_id', how='left')
df_productos.rename(columns={'cat1': 'product_category'}, inplace=True)
df_productos
```

```python
df_producto_estrella = df_productos[df_productos['product_id'] == 20001]
```

```python
# Find all unique values for 'periodo' and 'product_id'
unique_periodos = df['periodo'].unique()
unique_product_ids = df['product_id'].unique()
unique_customer_ids = df['customer_id'].unique()
len(unique_periodos), len(unique_product_ids), len(unique_customer_ids)
```

```python
print("Shape antes: ", df_productos.shape)

from itertools import product

# Create all possible combinations of 'periodo' and 'product_id'
all_combinations = list(product(unique_periodos, unique_product_ids))

# Convert the combinations into a DataFrame
all_combinations_df = pd.DataFrame(all_combinations, columns=['periodo', 'product_id'])

# Merge with the existing DataFrame to find missing rows
combined_df = all_combinations_df.merge(df_productos, on=['periodo', 'product_id'], how='left')

# Identify the missing rows by looking for NaNs
missing_rows = combined_df[combined_df['cust_request_qty'].isna()]

# For the missing rows, set 'cust_request_qty', 'cust_request_tn', and 'tn' to 0
# We also need to decide how to handle 'product_category'. For now, we'll set it as 'Unknown'
missing_rows.loc[:, 'cust_request_qty'] = 0
missing_rows.loc[:, 'cust_request_tn'] = 0
missing_rows.loc[:, 'tn'] = 0

# Look for other row with this product_id and asign the category
for index, row in missing_rows.iterrows():
    product_id = row['product_id']
    product_category = df_productos[df_productos['product_id'] == product_id]['product_category'].values[0]
    missing_rows.loc[index, 'product_category'] = product_category

# Now we append the missing rows to the original DataFrame
# convert to pd.concat: complete_df = df_productos.append(missing_rows[df_productos.columns], ignore_index=True)
df_productos = pd.concat([df_productos, missing_rows[df_productos.columns]], ignore_index=True)

# Return the number of new rows added and the path to the new file
num_new_rows = len(missing_rows)
print(num_new_rows)

print("Shape DespuÃ©s: ", df_productos.shape)
```
