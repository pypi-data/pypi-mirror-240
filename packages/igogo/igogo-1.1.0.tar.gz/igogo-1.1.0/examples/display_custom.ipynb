```python
import matplotlib.pyplot as plt
import numpy as np

%load_ext igogo
import igogo
```

```python
def f(x):
    return np.sin(x)
```

```python
@igogo.job(warn_rewrite=False, displays=2)
def run():
    for i in range(1, 100):
        x = np.linspace(0, i / 10, 100)
        fig = plt.figure()
        plt.plot(
            x,
            f(x)
        )
        igogo.display(fig)
        
        fig = plt.figure()
        plt.scatter(
            x,
            f(x)
        )
        igogo.display(fig)
        igogo.sleep(0.02)
    return "success"
    

result = run()
```
