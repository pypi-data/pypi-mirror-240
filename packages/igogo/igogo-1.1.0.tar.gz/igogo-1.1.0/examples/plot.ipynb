```python
import numpy as np
import matplotlib.pyplot as plt
import igogo

def experiment(name, f, i):
    x = np.linspace(0.01, i, 1000)
    fig, ax = plt.subplots(figsize=(5, 1))
    ax.plot(x, f(x), c='r')
    ax.set_title(name)
```

```python
@igogo.job(warn_rewrite=False)
def hello_world_matplotlib(name, f):
    for i in range(2, 10):
        experiment(name, f, i)
        igogo.sleep(0.5)  
    return name
```

```python
hello_world_matplotlib("cos", np.cos)
hello_world_matplotlib("sin", np.sin)
None
```

```python
experiment("atan", np.arctan, 10)
```
