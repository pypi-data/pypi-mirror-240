```python
import matplotlib.pyplot as plt
import numpy as np

%load_ext igogo
import igogo
```

```python
def f(x):
    return np.sin(x)
```

```python
@igogo.job(warn_rewrite=False)
def run():
    for i in range(1, 100):
        x = np.linspace(0, i / 10, 100)
        plt.plot(
            x,
            f(x)
        )
        igogo.sleep(0.05)
    return "success"
    

result = run()
```
