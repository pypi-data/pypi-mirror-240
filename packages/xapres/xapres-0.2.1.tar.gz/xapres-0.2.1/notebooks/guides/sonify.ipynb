```python
import sys
sys.path.append("../../../xapres/")
import xapres as xa
```

```python
from_zarr = xa.load.load_zarr()
from_zarr
```

```python
one_chirp = from_zarr.chirp.isel(time=300, chirp_num=0, attenuator_setting_pair=0).load()
```

```python
one_chirp.sonify(save = True)
```

```python
from_DAT_unattended = xa.load.generate_xarray(directory='../../data/sample/single_dat_file/', 
            file_numbers_to_process = [0], 
            bursts_to_process=[0],
            )
from_DAT_unattended
```

```python
from_DAT_unattended.chirp.isel(time=0, chirp_num=0, attenuator_setting_pair=0).sonify(save = True)
```
